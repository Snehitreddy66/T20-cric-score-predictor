```python
import pickle
import pandas as pd
import numpy as np
```

```python
df = pickle.load(open("dataset_level2.pkl","rb"))
```

```python
df
```

```python
# we want 8 columns


# batting_team
# bowling team
# city
# current score
# ball left
# wickets left
# current rr
# last five
```

```python
df.isnull().sum()
```

```python
df["city"].isnull()
```

```python
df[df['city'].isnull()]['venue'].value_counts()
```

```python
cities = np.where(df['city'].isnull(),df['venue'].str.split().apply(lambda x:x[0]),df['city'])

#the code will return as venue name where city is null and city name where city is not null
```

```python
cities    
```

```python
df['city'] = cities
```

```python
df.isnull().sum()
# now there are no missing values
```

```python
df.drop(columns=['venue'],inplace=True)
#now I drop venue column
```

```python
df
```

```python
eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist()
# I get the cities where min 5 matches are played 
```

```python
df = df[df['city'].isin(eligible_cities)]
```

```python
df
```

```python
df['current_score'] = df.groupby('match_id').cumsum()['runs']
# calling the Runs column by using cumulativesum and getting in new column current score
```

```python
df
```

```python
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
df

# seperated the over and ball column
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
df

#counting the balls in over multiplying by 6
```

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
df

# substracted the balls from 120 due to extra balls for correct result
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
df['player_dismissed'] = df['player_dismissed'].astype('int')
df['player_dismissed'] = df.groupby('match_id').cumsum()['player_dismissed']
df['wickets_left'] = 10 - df['player_dismissed']
```

```python
df    # with this I can see how many wickets left after each and every ball
```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
 # I want current runrate so I will curren score *6/ balls 
```

```python
df
```

```python
groups = df.groupby('match_id')

match_ids = df['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30).sum()['runs'].values.tolist())
    
    
# i devided the dataset on the basis of match id  into groups
# with this the data will devide after each match
#then i wrote loop function on the match id's
#and get the dataframe of the particular match id then i applied rolling function from pandas to 30[5(overs)*6(balls)]
#applied sum on runs column.
```

```python
last_five      # I cant get first 30 balls as null because I will get values if min 30 balls are played in each mtch
```

```python
df['last_five'] = last_five
# storing in df column
```

```python
# getting Final score by sum on runs column on the basis of match id and groupby
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
final_df
```

```python
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
```

```python
final_df
```

```python
# dropping missing values 
final_df.isnull().sum()
```

```python
#dropping that missing values
final_df.dropna(inplace=True)
```

```python
final_df = final_df.sample(final_df.shape[0])
# shuffling the data
```

```python
final_df
```

```python
final_df
```

```python
#spliting the train Test and building the model

X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
#importing dependancies from sklearn

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
# make column transfer because three columns are string I have to one hot encode them


trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
# make sklearn pipeline where in 1st step data will transform on the basis of above code 
# in 2nd step all values will become into scale
# in 3rd step I will implement xgboost regressor with 


pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
# train the model and predict.

pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
# now I will export this model for website creation in pycharm and deploy it

pickle.dump(pipe,open('pipe.pkl','wb'))
```
